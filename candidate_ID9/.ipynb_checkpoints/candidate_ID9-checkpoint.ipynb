```python
#Exercice 4
from functools import wraps # Le module qui decore une focntion en la conservant


def affichage():
    def decorateur(fonction_style):
        @wraps(fonction_style)
        def fonction_decoree(*args, **kwargs):
            resultat = fonction_style(*args, **kwargs)
            s=str(resultat)
            try:
                if (kwargs['g'] ): #teste si g est vrai
                    s="*"+s+"*"
                if (kwargs['i'] ): #teste si i est vrai
                        s="_"+s+"_"
                if (kwargs['b'] ): #teste si b est vrai
                        s="~"+s+"~"
            except KeyError: #si un argument manque
                try:
                    if (kwargs['i'] ): #teste les autres
                        s="_"+s+"_"
                    if (kwargs['b'] ):
                            s="~"+s+"~"
                except KeyError: # au cas ou l'un manque encore
                    try:
                        if (kwargs['b'] ): #tes
                            s="~"+s+"~"
                    except KeyError:
                        pass   
            return s # retourne le texte decore
        return fonction_decoree
    return decorateur


@affichage()
def style(text, g=False, i=False, b=False):
    return text

print(style("the",b=True,g=True))
input()
```

## Exercice 9 (25 points): Chiffrement de César

```python
#Exercice 8

 
clair=list("ABCDEFGHIJKLMNOPQRSTUVWXYZ") #alphabet clair
cypher =list("XYZABCDEFGHIJKLMNOPQRSTUVW") #alphabet cypher
        
def encodage(text):
    chiffrement=dict(zip(clair, cypher)) #creation d'un dictionnaire cle dans alphabet clair et valeur dans alphabet cypher
    text=text.split() #text en liste
    for elt in text:
        elt=list(elt)
        for i in range(len(elt)):
            elt[i]=chiffrement[elt[i]]
        elt="".join(elt) #recoller la liste
        print(elt,end=" ")
    
def decodage(text):
    dechiffrement=dict(zip(cypher, clair))#creation d'un dictionnaire cle dans alphabet cypher et valeur dans alphabet clair
    text=text.split()
    for elt in text:
        elt=list(elt)
        for i in range(len(elt)):
            elt[i]=dechiffrement[elt[i]]
        elt="".join(elt)
        print(elt,end=" ")

decodage("LE BON")
encodage("LE BON")
input('')
```
